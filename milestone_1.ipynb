## TASK 2

* Load a 512x512 image
* Initialize with random rgb values 
* Tranform the image into a grayscale image
* Display it

#### Convert `TASK 1` Image to corresponding grayscale image 
(without using any inbuilt function of `opencv`)

```python
## Hint: use the variable from TASK 1

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
# read the image file as a numpy array
image = np.array(Image.open('random_image.png'))

# get the height and width of the image
height, width = image.shape[:2]

# create a new array to hold the grayscale image
grayscale_image = np.zeros((height, width), dtype=np.uint8)

# loop through each pixel in the image and convert it to grayscale
for i in range(height):
    for j in range(width):
        r = image[i,j,0]
        g=image[i,j,1]
        b=image[i,j,2]
        grayscale_pixel = int(0.2989 * r + 0.5870 * g + 0.1140 * b)
        grayscale_image[i,j] = grayscale_pixel
```

#### Display the grayscale image

```python
## Hint: use matplotlib
plt.imshow(grayscale_image, cmap='gray')
plt.show()


# save the grayscale image as a file in the current directory
grayscale_image_obj = Image.fromarray(grayscale_image)
grayscale_image_obj.save('grayscale_image.png')
```
